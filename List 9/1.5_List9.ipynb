```python
data_string = '''R2                      0.146          0.182  '''
rest, unrest =np.float_(re.findall(r'([-+]?\d+.\d+)', data_string))
```

```python
F_obs = (unrest - rest)/(1-unrest)*(935-8)/4
F_obs.round(2)
```

$$
H0: \beta_{urban} = \beta_{age*urban} = \beta_{IQ*urban} =  \beta_{south*urban}
$$
$$
H_1: \beta_{urban}^2 + \beta_{age*urban}^2 + \beta_{IQ*urban}^2 + \beta_{south*urban}^2 > 0 
$$

```python
alpha = 0.1
F_cr = stats.f.ppf(1-alpha,4,935-8)
F_cr
```

$$
