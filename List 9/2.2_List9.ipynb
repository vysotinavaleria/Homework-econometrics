```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
import seaborn as sns
import re
import scipy.stats as stats
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/artamonoff/Econometrica/master/python-notebooks/data-csv/diamonds.csv')
# Метод .loc[] фильтрует данные таблицы по строкам и столбцам.
# Метод  "df.loc[:,['carat']]" делает срез данных и означает,
# что из датафрейма выбираются все строки(это делается с помощью знака ":") столбца "carat".
df_all = pd.get_dummies(df.loc[:,['carat','color','cut','price']])
df_all
```

```python
df_all.rename(columns = {'cut_Very Good':'cut_Very_Good'}, inplace = True)
df_all
```

```python
model_d = smf.ols(formula='np.log(price)~carat+I(carat**2)+color_E+color_F+color_G+color_H+color_I+color_J+cut_Good+cut_Very_Good+cut_Premium+cut_Ideal',data=df_all).fit()
model_d.params
```

```python
model_d.f_test('color_E=color_F=color_G=color_H=color_I=color_J=0')
```

```python
stats.f.ppf(1-0.05,6,len(df_all)-6).round(2)
```
