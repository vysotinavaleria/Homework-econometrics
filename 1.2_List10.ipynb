```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import statsmodels.formula.api as smf
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/ryupepa/Econometrica/master/python-notebooks/data-csv/sleep75.csv')
df
```

```python
model = smf.ols(data=df, formula='sleep~totwrk+south+totwrk*south+age+age*age+male+smsa').fit()
hat_output = model.predict()
model.predict()
```

```python
df['slp2'] =model.predict()**2
df['slp3'] =model.predict()**3
df['slp4'] =model.predict()**4
df
```

```python
model2 = smf.ols(data=df, formula='sleep~totwrk+south+totwrk*south+age+age*age+male+smsa+slp2+slp3+slp4').fit()
model2.params
```

```python
R2=model2.rsquared
R2_res = model.rsquared
```

```python
F_obs=((R2 - R2_res)/(1-R2))*((len(df)-12)/3)
F_obs
```

```python
alpha = 0.05
stats.f.ppf(1-alpha,3,len(df)-12).round(2)
```
