```python
import numpy as np
import pandas as pd
import io
import statsmodels.formula.api as smf
```

```python
data = '''output capital labour
9.25    2.61    184  
3.66    1.32     91  
28.78   22.09   426  
4.12    10.74    72  
2.89    1.16     46 '''
```

```python
df = pd.read_csv(io.StringIO(data), sep='\s+')
df
```

```python
# Переименуем столбцы, так как log(output) интерпреитуется как функция логарифма от outpu
df =df.rename(columns={"output": "LOGoutput", "capital": "LOGcapital", "labour":'LOGlabour'})
df
```

```python
df ['pred'] = -1.71+0.21 *np.log(df['LOGcapital']) +0.71 *np.log(df['LOGlabour'])
df
```

```python
df['residual']=np.log(df['LOGoutput'])-df['pred']
df
```
