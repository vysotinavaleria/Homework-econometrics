H0: \beta_age = \beta_age^2 = 0

H1: \beta_age^2 + \beta_age^4 > 0

```python
import pandas as pd
import numpy as np
import io
import statsmodels.formula.api as smf
import scipy.stats as stats
import re
import scipy
```

```python
unrest, rest = np.float_(re.findall(r'([-+]?\d+.\d+)','R2                      0.128          0.122'))
print (unrest, rest)
```

```python
F_obs = (unrest - rest)/(1-unrest)*(706 - 6)/2
F_obs
```

```python
alpha = 0.01
F_cr = stats.f.ppf(1-alpha, 2, 706-6)
F_cr
```

Fobs < Fcr => данные согласуются с гипотезой H0 и совместное влияние не значимо

```python
# проверка значимости коэффициентов
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/artamonoff/Econometrica/master/python-notebooks/data-csv/sleep75.csv')
df
```

```python
df['age^2']  = df['age']*df['age']
df
```

```python
sleep_eq = smf.ols(formula='sleep~totwrk+age+age^2+male+south', data=df).fit()
sleep_eq.params.round(3)
```

```python
model = smf.ols(formula='sleep~totwrk+age+age^2+male+south', data=df).fit()
model.summary(alpha=0.01).tables[1]
```

```python
df = pd.read_html(model.summary().tables[1].as_html(), header=0, index_col=0)[0] 
df
```

```python
df['tvalues']  = df['coef']/df['std err']
df
```

```python
alpha = 0.01
print('Уровень значимости alpha: ', alpha)
DF = int(model.df_resid) 
t_crit = scipy.stats.t.ppf(1-alpha/2, DF)
print('Критическое значение t_crit: ', t_crit.round(3))
```

```python
df['znachimosty'] = df.apply(lambda x : 'znachim' if np.abs(x['tvalues']) > t_crit else 'ne znachim', axis = 1 )
df
```
