```python
import numpy as np
import pandas as pd
import io
import statsmodels.formula.api as smf
import scipy.stats as stats
import scipy
import seaborn as sns
import re
import matplotlib.pyplot as plt
```

### Загрузим датафрейм

```python
df = pd.read_csv('https://raw.githubusercontent.com/vysotinavaleria/Homework-econometrics/main/penguins_raw.csv')
df
```

### Создадим регрессию по методу наименших квадратов

#### Формула
$$
Body-Mass = \beta_0 + \beta_1*Sex+\beta_2*Culmen-Length+\beta_3*Flipper-Length+\beta_4*Species
$$

```python
df.rename(columns = {'Culmen Length (mm)':'Culmen_Length', 'Flipper Length (mm)':'Flipper_Length', 'Body Mass (g)':'Body_Mass', 'Culmen Depth (mm)':'Culmen_Depth'}, inplace = True)
df
```

Присвоим значениям некоторых столбцов цифровые значения, чтобы использовать их при построении регресси

```python
df = df.replace({'Sex':{'FEMALE':1, 'MALE':0}})
df = df.replace({'Species':{'Adelie Penguin (Pygoscelis adeliae)':0, 'Gentoo penguin (Pygoscelis papua)':1, 'Chinstrap penguin (Pygoscelis antarctica)':2}})
df
```

```python
penguin = smf.ols(formula='Body_Mass ~Sex+Culmen_Length+Flipper_Length+Species', data=df).fit()
penguin.params.round(2)
```

### Значение полученных коэффициентов:
- У самцов пингвинов вес выше на 249.35 грамм, чем у самок при прочих равных
- Если длина клюва пингвина увеличится на 1 мм, его вес при этом увеличится на 30.95 грамм при прочих равных
- При увеличенаделии длины крыла пингвина на 1 мм, его вес увеличится на 44.63 гр при прочих равных
- У пингвинов Адели самый большой вес - на 244.82 гр больше, чем у папунских пингвинов, чей вес на  244.82 гр больше, чем у арктических пингвинов - они самые легкие

## Определим надежность построенной модели с помощью коэффициента детерминации

```python
TSS =penguin.centered_tss.round(3)
```

```python
RSS = penguin.ssr.round(3)
```

```python
R2 = RSS/TSS
R2
```

```python
# Значение коэффециента детерминации ближе к 0,а не к 1, регрессия не надежна
```

### Стандартная ошибка
$$
s^2 = RSS/(n-m)
$$
n - количество наблюдений

m - количество регрессоров + 1

```python
s2 = RSS/(344 - 5)
s2
```

### Исправленный коэффициент детерминации
$$
R^2_{adj} = 1 - (1-R^2)*(n-1)/(n-m)
$$

```python
R2adj = 1 - (1 - R2)*(344-1)/(344 - 5)
R2adj
```

```python
# Исправленный коэффициент детерминаии еще меньше,регрессия не точная
```

## Проведем RESET-test

```python
df = df.dropna ()
df
```

```python
model3 = smf.ols(data=df, formula='Body_Mass~Sex+Culmen_Length+Flipper_Length+Species').fit()
model3.predict()
```

```python
df['bm2'] =model3.predict()**2
df['bm3'] =model3.predict()**3
df['bm4'] =model3.predict()**4
df
```

```python
model4 = smf.ols(data=df, formula='Body_Mass~Sex+Culmen_Length+Flipper_Length+Species+bm2+bm3+bm4').fit()
model4.params
```

```python
R4=model4.rsquared
R4_res = model3.rsquared
```

```python
F_obs=((R4 - R4_res)/(1-R4))*((len(df)-7)/3)
F_obs
```

```python
alpha = 0.05
stats.f.ppf(1-alpha,3,len(df)-7).round(2)
```

$$
F_{obs}<F_{cr}
$$
Данные согласуются с гипотезей Н0, не нужно менять степень зависимости, все необходимые переменные присутствуют
