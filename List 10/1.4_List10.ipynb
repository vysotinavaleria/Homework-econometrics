```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import statsmodels.formula.api as smf
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/ryupepa/Econometrica/master/python-notebooks/data-csv/Labour.csv')
df
```

```python
model = smf.ols(data=df, formula='np.log(output)~np.log(capital)+np.log(labour)+np.log(wage)').fit()
hat_output = model.predict()
model.predict()
```

```python
model.params
```

```python
# Дополним датасет предсказанными знчениями переменной ouput соответсвующей степени
df['out2'] =model.predict()**2
df['out3'] =model.predict()**3
df['out4'] =model.predict()**4
df
```

```python
model2 = smf.ols(data=df, formula='np.log(output)~np.log(capital)+np.log(labour)+np.log(wage)+np.log(out2)+np.log(out3)+np.log(out4)').fit()
model2.params
```

```python
R2=model2.rsquared
R2_res = model.rsquared
```

```python
F_obs=((R2 - R2_res)/(1-R2))*((len(df)-7)/3)
F_obs
```

```python
alpha = 0.05
stats.f.ppf(1-alpha,3,len(df)-7).round(2)
```
